```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train_data.csv', index_col = 0)
target = pd.read_csv('train_target.csv', index_col = 0)
```

```python
train.head()
```

```python
target.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.5, random_state=42)
```

```python
X_train.head()
```

```python
X_train.to_csv('train.csv', index=False)
y_train.to_csv('target.csv', index=False)
X_test.to_csv('test.csv', index=False)
```
